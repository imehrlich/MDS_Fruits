```python
import itertools
import pandas as pd
import numpy as np

wcsData = pd.read_csv('wcsDataLong.csv')
wcsData.head()
```

```python
all_pairs = pd.DataFrame() 

for k in range(1, max(wcsData['language'])+1):
    lnk = wcsData[(wcsData['language'] == k)]

    for j in range(1, max(lnk['speaker'])+1):

        lnkspkj = lnk[(lnk['speaker'] == j)]
        lnkspkj.index = range(0,330)


        pairs = pd.DataFrame(itertools.combinations(range(1,331), 2))
        pairs = pairs.rename(columns={0:"chipID1", 1:"chipID2"})
        pairs['same'] = np.zeros(len(pairs.index), dtype=int)
        pairs['language'] = k
        pairs['speaker'] = j

        for i in range(len(pairs.index)):

            chip1 = pairs['chipID1'][i]
            chip2 = pairs['chipID2'][i]

            term1 = lnkspkj[lnkspkj['chipID'] == chip1]['term'][chip1-1]
            term2 = lnkspkj[lnkspkj['chipID'] == chip2]['term'][chip2-1]

            if term1 == term2:
                pairs.iat[i, 2] = 1

        all_pairs = pd.concat([all_pairs, pairs])
        
all_pairs.to_csv('pairs.csv', index=False)
```
